Mathematical Operations

1) Scalar Operations

```python
x = torch.rand(2,2)
x
```

```python
# addition
x + 2
# substraction
x - 2
# multiplication
x * 3
# division
x / 3
# int division
(x * 100)//3
# mod
((x * 100)//3)%2
# power
x**2
```

2) Element wise operation

```python
a = torch.rand(2,3)
b = torch.rand(2,3)

print(a)
print(b)
```

```python
# add
a + b
# sub
a - b
# multiply
a * b
# division
a / b
# power
a ** b
# mod
a % b
```

```python
c = torch.tensor([1, -2, 3, -4])
```

```python
#abs
torch.abs(c)
```

```python
d = torch.tensor([1.9, 2.3, 3.7, 4.4])
```

```python
# round
torch.round(d)
```

```python
# ceil
torch.ceil(d)
```

```python
#floor
torch.floor(d)
```

```python
# clamp
torch.clamp(d, min = 2, max = 3)
```

3) Reduction operation

```python
e = torch.randint(size=(2,3), low=0, high=10, dtype=torch.float32)
e
```

```python
# sum
torch.sum(e)
```

```python
# sum along columns
torch.sum(e, dim=0)
```

```python
# sum along rows
torch.sum(e, dim=1)
```

```python
# mean
torch.mean(e) # net total mean
# mean along col
torch.mean(e, dim=0)
```

```python
#median
torch.median(e)
```

```python
# max and min
torch.max(e)
torch.min(e)
```

```python
# argmax
torch.argmax(e)
```

```python
torch.argmin(e)
```

4) Matrix operations

```python
f = torch.randint(size=(2,3), low=0, high=10)
g = torch.randint(size=(3,2), low=0, high=10)

print(f)
print(g)
```

```python
# matrix multiplication
torch.matmul(f, g)
```

```python
vector1 = torch.tensor([1, 2])
vector2 = torch.tensor([3, 4])

# dot product
torch.dot(vector1, vector2)
```

```python
# transpose
torch.transpose(f, 0, 1) #swap dimension 0 and 1
```

```python
h = torch.randint(size=(3,3), low=0, high=10, dtype=torch.float32)
h
```

```python
# determinant
torch.det(h)
```

```python
# inverse
torch.inverse(h)
```

5) Comparison Operations

```python
k = torch.randint(size=(2,3), low=0, high=10, dtype=torch.float32)
k
```

```python
# log
torch.log(k)
```

```python
# exp
torch.exp(k)
```

```python
# sqrt
torch.sqrt(k)
```

```python
# sigmoid
torch.sigmoid(k)
```

```python
# softmax
torch.softmax(k, dim=0) # across comumns
```

```python
# relu
torch.relu(k)
```
