```python
# Create model class
import torch
import torch.nn as nn

class Model(nn.Module):       # Model is a subclass of nn.Module

  def __init__(self, num_features):    # initializer : runs when u create a model object

    super().__init__()                 # Also run the parent class (nn.Module)’s initializer.
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):

    out = self.linear(features)
    out = self.sigmoid(out)

    return out
```

```python
class Model(nn.Module):

  def __init__(self, num_features):

    super().__init__()
    self.linear = nn.Linear(num_features, 3)
    self.relu = nn.ReLU()
    self.linear2 = nn.Linear(3, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):

    out = self.linear(features)
    out = self.relu(out)
    out = self.linear2(out)
    out = self.sigmoid(out)

    return out
```

```python
# create model class  # Sequential containers
import torch
import torch.nn as nn

class Model(nn.Module):

  def __init__(self, num_features):

    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(num_features, 3),
        nn.ReLU(),
        nn.Linear(3, 1),
        nn.Sigmoid()
    )

  def forward(self, features):

    out = self.network(features)

    return out
```

```python
# create dataset
features = torch.rand(10,5)

# create model
model = Model(features.shape[1])

# forward pass call
model(features)  # same as model.forward(features)
```
